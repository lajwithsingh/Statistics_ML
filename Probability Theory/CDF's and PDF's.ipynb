```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
##example of using log-normal distribution

#the variable to evaluate function on
x = np.linspace(0, 5, 10001)

p1 = stats.lognorm.pdf(x, 1)
c1 = stats.lognorm.cdf(x, 1)

p2 = stats.lognorm.pdf(x, .1)
c2 = stats.lognorm.cdf(x, .1)
```

```python
#draw the pdfs
fig, ax = plt.subplots(2, 1, figsize=(4,7))

ax[0].plot(x, p1/sum(p1), x , p2/sum(p2))
ax[0].set_ylabel('Probability')
ax[0].set_title("pdf(x)")

#draw the cdfs
ax[1].plot(x, c1, x, c2)
ax[1].set_ylabel("Probability")
ax[1].set_title("cdf(x)")
plt.show()
```

```python
c1x = np.cumsum(p1)
plt.plot(x, c1x, '--')
plt.show()
```
