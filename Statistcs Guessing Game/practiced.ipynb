Importing The Important Libraries

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats 
```

```python
#Decide The Average for 2 groups
average_group1 = 40
average_group2 = 38

#Decide the Sample Size for 2 groups
sample_size1 = 40
sample_size2 = 40

#Decide the Standard Deviation for 2 groups
std = 5.6
```

```python
#Creating The Two Random Dataset
data_group1 = np.random.randn(sample_size1) * std + average_group1 
data_group2 = np.random.randn(sample_size2) * std + average_group2 

#Combining the Two groups
ns = [sample_size1, sample_size2]

#finding the Data limits for y axis 

datalims = [np.min(np.hstack((data_group1, data_group2))), np.max(np.hstack((data_group1, data_group2)))]
```

```python
#Visualizing the Two Plots

fig, ax  = plt.subplots(1,2,figsize=(6,4))

ax[0].violinplot(data_group1)
ax[0].plot(1 + np.random.randn(sample_size1)/10, data_group1, "ko")
ax[0].set_ylim(datalims)
ax[0].axis('off')

ax[1].violinplot(data_group2)
ax[1].plot(1 + np.random.randn(sample_size2)/10, data_group2, "ko")
ax[1].set_ylim(datalims)
ax[1].axis('off')

#Testing the Two Groups 
t, p = stats.ttest_ind(data_group1, data_group2)

#options for title 
option = [""," Not"]

#putting the title
plt.title("The Two Groups Are {} Statistically Significant \nWith T Value {} and P Value {}".format(option[int(p>0.05)],round(t,2), p))
```
