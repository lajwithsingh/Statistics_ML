<h1>Inter Quartile Range</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n = 1000
data = np.random.randn(1000) ** 2

fig, ax = plt.subplots(1,2, figsize = (15, 5))

ax[0].plot(data, 'p')

ax[1].hist(data, rwidth = 0.95)
ax[1].set_title('Histogram Showing the Distribution')
```

```python
dataR = stats.rankdata(data)/n

fig, ax = plt.subplots(1,2, figsize = (15, 5))

ax[0].plot(dataR, 'rp')

ax[1].plot(data,dataR,'rp')
```

```python
dataR
```

```python
#finding the value closes to quartile 1 and quartile 3

q1 = np.argmin((dataR-.25) ** 2)
q3 = np.argmin((dataR-.75) ** 2)

iqr_vals = data[[q1, q3]]

#IQR Value 1
iqr_value1 = iqr_vals[1] - iqr_vals[0]

#IQR Value 2
iqr_value2 = stats.iqr(data)

print(iqr_value1, iqr_value2)
```
