<h1>Violin Plot</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Generating Data
n = 1000
thresh = 5

data = np.exp(np.random.randn(n))
data[data>thresh] = thresh + np.random.randn(sum(data>thresh))*.1

plt.hist(data,bins = 'fd')
plt.show()
```

```python
#Violin Plot Using Matplotlib

plt.violinplot(data)
plt.show()
```

```python
#Violin Plot Using Seaborn

sns.violinplot(data = data, kind='violin')
```
