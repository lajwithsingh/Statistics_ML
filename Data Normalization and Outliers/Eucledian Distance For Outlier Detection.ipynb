<h1>Eucledian Distance For Outlier Detection</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Create Some Data

N = 40

# two-dimensional data

d1 = np.exp(-abs(np.random.rand(N)*3))
d2 = np.exp(-abs(np.random.rand(N)*5))
datamean = [np.mean(d1), np.mean(d2)]

#compute distance of each point to the mean

ds = np.zeros(N)
for i in range(N):
    ds[i] = np.sqrt((d1[i]-datamean[0]) ** 2 + (d2[i]-datamean[1]) ** 2)
    
# convert to z (don't need the original data)

ds = (ds-np.mean(ds))/np.std(ds)

#plot the data

fig, ax = plt.subplots(1,2, figsize=(12, 6))
ax[0].scatter(d1, d2)
ax[0].set_ylabel("Y Variable")
ax[0].set_xlabel("X Variable")
ax[0].plot(datamean[0], datamean[1], 'gp', markersize = 15)

ax[1].scatter(np.arange(N), ds)
ax[0].set_title("Z Scored Converted Data")
ax[0].set_xlabel("Index")
ax[0].set_ylabel("Z Scores")
```

```python
distancethresh = 2

#offending points
oidx = np.where(ds>distancethresh)[0]

#marking on visual
ax[0].plot(d1[oidx],d2[oidx],'rx',markersize = 12)
ax[1].plot(oidx, ds[oidx], 'rx', markersize = 12)

fig
```
