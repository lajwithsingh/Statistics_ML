<h1>Un Supervised Learning Z Vs Modified Z</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels import robust
from scipy import stats
```

```python
N = 200

data1 = np.random.randn(N)

#Z-score
zScoreStandardized = (data1 - np.mean(data1))/(np.std(data1))

#Modified Z Score

median = np.median(data1)
mad = robust.mad(data1)

modifiedZScoreStandardized = 0.6745*(data1 - median) / mad
```

```python
zscorethresh = 3

fig, ax = plt.subplots(2, 1, figsize = (8, 6))

ax[0].plot(zScoreStandardized , 'k^', markerfacecolor = 'w', markersize = 12)
ax[0].set_xlabel('Data Index')
ax[0].set_ylabel('Orig Scale')

ax[1].plot(modifiedZScoreStandardized  , 'k^', markerfacecolor = 'w', markersize = 12)
ax[1].set_xlabel('Data Index')
ax[1].set_ylabel('Orig Scale')

plt.show()
```

```python
plt.plot(zScoreStandardized, modifiedZScoreStandardized, 'r*')
```

```python
plt.hist(data1,bins = 50, rwidth = 0.5)
plt.show()
```

```python
N = 200

data2 = np.random.randn(N) ** 4

#Z-score
zScoreStandardized2 = (data2 - np.mean(data2))/(np.std(data2))

#Modified Z Score

median = np.median(data2)
mad = robust.mad(data2)

modifiedZScoreStandardized2 = 0.6745*(data2 - median) / mad
```

```python
def outliers_modified_z_score(ys):
    threshold = 3.5

    median_y = np.median(ys)
    median_absolute_deviation_y = np.median([np.abs(y - median_y) for y in ys])
    modified_z_scores = [0.6745 * (y - median_y) / median_absolute_deviation_y for y in ys]
    return modified_z_scores
```

```python
zscorethresh = 3

fig, ax = plt.subplots(3, 1, figsize = (8, 6))

ax[0].plot(zScoreStandardized2 , 'k^', markerfacecolor = 'w', markersize = 12)
ax[0].set_xlabel('Data Index')
ax[0].set_ylabel('Orig Scale')

ax[1].plot(modifiedZScoreStandardized2  , 'k^', markerfacecolor = 'w', markersize = 12)
ax[1].set_xlabel('Data Index')
ax[1].set_ylabel('Orig Scale')

ax[1].plot(outliers_modified_z_score(data2)  , 'k^', markerfacecolor = 'w', markersize = 12)
ax[1].set_xlabel('Data Index')
ax[1].set_ylabel('Orig Scale')
plt.show()
```

```python
plt.plot(zScoreStandardized2, modifiedZScoreStandardized2, 'r*')
```
