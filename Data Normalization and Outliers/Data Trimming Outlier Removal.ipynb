<h1>Data Trimming Outlier Removal</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##create some data

N = 40

data = np.random.randn(N, 1)
data[data<-2] = -data[data<-2] ** 2
data[data>2] = data[data>2] ** 2

#Also Need Mean Centered Data

dataMC = data - np.mean(data)

#plotting the data

fig, ax = plt.subplots(1, 1, figsize = (10, 10))
ax.plot(data, 'k^', markersize = 12, markerfacecolor = 'y')
ax.set_yticks([])
ax.set_xticks([])
ax.set_xlabel("Index")
ax.set_ylabel("Data")
plt.show()
```

```python
#option1 how much percent

trimpercent = 5

#Identify the Cut Off
datacutoff = np.percentile(abs(dataMC), 100 - trimpercent)

#find the Exceedence value
data2cut = np.where(abs(dataMC) > datacutoff)[0]

ax.plot(data2cut, data[data2cut], 'rx', markersize = 15)
fig
```

```python
##option2 : remove k most extreme values

k2remove = 3
datasortIdx = np.argsort(abs(dataMC), axis = 0)[::-1]
data2cut = np.squeeze(datasortIdx[:k2remove])
```

```python
ax.plot(data2cut, data[data2cut], 'go', markersize = 15, alpha = 0.5)
fig
```
