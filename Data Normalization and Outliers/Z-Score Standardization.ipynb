<h1>Z Score Standardization</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
normally_distributed = np.random.randn(1000) * 19 + 27

plt.hist(normally_distributed) 
```

```python
gaussian = np.random.normal(18, 4, 1000)

plt.hist(gaussian)
```

```python
mean =  np.mean(gaussian)
std = np.std(gaussian)

print(mean)
print(std)
```

```python
#Z Score - Standardization

standardized = (gaussian - mean)/std

plt.hist(standardized,bins = 40, rwidth = 0.5)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
standardized2 = scaler.fit_transform(gaussian.reshape(-1, 1))

plt.hist(standardized2, bins = 40, rwidth = 0.5)
plt.show()
```

```python
from scipy import stats

stats.probplot(gaussian, plot = plt)
plt.show()
```
