<h1>Histogram's In Python</h1>
<p>In this notbook we will discuss how to plot the histogram using python libraries</p>

**Importing the libraries**

```python
import pandas as pd   
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
```

**Generate Random Distribution**

```python
uniform = np.random.uniform(size = 1000)
exponential = np.random.exponential(size = 1000)
normal = np.random.normal(size = 1000)


data = pd.DataFrame({"Uniform":uniform, "Exponential":exponential,"Normal":normal})
data.head()
```

**Plotting Histogram Using Matplotlib**

```python
plt.style.use('fivethirtyeight')


colors = ["#f765A3", "#f76543" ,"#a15589"]
widths = [0.005, 0.05, 0.05]

fig, ax = plt.subplots(1, 3, figsize=(24, 8))

for i in range(3):
    ax[i].hist(data[data.columns[i]], bins = 100,width = widths[i], color = colors[i])
    ax[i].set_title(f"{data.columns[i]} Distribution  Plot")
```

**Histogram Using Seaborn**

```python
fig, ax = plt.subplots(1, 3, figsize = (24, 8))

for i in range(3):
    sns.histplot(data = data[data.columns[i]], bins = 25 , color = colors[i], kde= True , ax = ax[i])
    ax[i].set_title(f"{data.columns[i]} Distribution Plot")
```

**Histogram Using Plotly**

```python
for i in range(3):
    fig = px.histogram(x = data[data.columns[i]], template = "plotly_white", nbins = 100, width = 600, title = f"{data.columns[i]} Distribution Plot")
    fig.update_traces(marker = {"line":{"width":3, "color":"white"}, "color":colors[i]})
    fig.show()
```
