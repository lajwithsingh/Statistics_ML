<h1>Box Plot Implementation in Python</h1>

**Import the Libraries**

```python
import pandas as pd
import plotly.express as px 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

**Importing the Dataset**

```python
data = px.data.gapminder()
data.head()
```

**Plotting Box Plot using Seaborn**

```python
fig  = plt.figure(figsize = (8, 4))
sns.boxplot(data = data, y = "lifeExp", x = "continent")
```

**Plotting Box Plot Using Matplotlib**

```python
fig = plt.figure(figsize = (8, 4))
plt.boxplot(data["lifeExp"])
```

```python
px.box(data_frame = data, x = "continent" , color = "continent", height = 400, width = 600 , y = "lifeExp" , template = "plotly_white" , title = "Box plot Showing the Distribution for Various Continents")
```
