<h1>Kernal Density Estimate Plot In Python</h1>
<p>In this Notebook We are going to Visualize The Distribution of Different Series of Numbers</p>

**Importing the Libraries**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns   
import plotly.express as px
```

**Generating Random Uniform & Normal Distribution**

```python
n = 10000

uniform = np.random.rand(n)
normal = np.random.randn(n)
exponential = np.random.exponential(size = n)

data = pd.DataFrame({"Uniform":uniform,"Normal" : normal, "Exponential":exponential})
data.head()
```

**Plotting KDE Plot Using Seaborn**

```python
plt.style.use('fivethirtyeight')

colors = ["#f765A3", "#f76543" ,"#a15589"]

fig, ax = plt.subplots(1, 3, figsize = (24, 8))

for i in range(3):
    sns.kdeplot(data[data.columns[i]], color = colors[i], ax = ax[i])
    ax[i].set_title(f"{data.columns[i]} Distribution Plot")

fig.show()
```
